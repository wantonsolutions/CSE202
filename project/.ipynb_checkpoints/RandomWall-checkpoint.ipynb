```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
width = 10
height = 30
```

```python
P = 0.4

wall = np.random.random((height, width))
plt.imshow(wall)
plt.colorbar()
plt.show()

plt.imshow(wall <= P)
plt.colorbar()
plt.show()
```

```python
for row in wall <= P:
    print(list(row.astype(int)))
```
